```python
ip = df.drop(['Student_ID', 'Academic_Performance'], axis=1)
op = df['Academic_Performance']
```

```python
# Train test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(ip,op, test_size=0.2)
```

```python
# Standard Scaler
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)  # Fit only on training data
x_test = sc.transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y1 = lr.predict(x_test)
```

```python
from sklearn.metrics import f1_score, recall_score, precision_score, accuracy_score
acc = accuracy_score(y1, y_test)
f1 = f1_score(y1, y_test, average='weighted')
pre = precision_score(y1, y_test, average='weighted')
rel = recall_score(y1, y_test, average='weighted')
print('Accuracy:', acc)
print('F1 Score:', f1)
print('Precision:', pre)
print('Recall:', rel)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm = confusion_matrix(y1,y_test)
cm
```

```python
cmd = ConfusionMatrixDisplay(cm)
cmd.plot()
plt.show()
```
